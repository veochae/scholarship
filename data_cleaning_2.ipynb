```python
import pandas as pd
import numpy as np
import yake
```

```python
us = pd.read_csv("./data/cleaned/us.csv")
dc = pd.read_csv("./data/cleaned/dc.csv")
```

```python
dict = {
    "min": [],
    "max": []
}

for k in us.salary:
    if k in [np.NaN]:
        dict['min'].append(np.NaN)
        dict['max'].append(np.NaN)

    elif "an hour" in k:
        t = k.replace(" an hour", '')
        t = t.replace(",", "")
        tt = np.char.split(t, "–").tolist()
        if len(tt) <2 :
            dict['min'].append(float(tt[0]) * 2080)
            dict['max'].append(np.NaN) 
        else:
            dict['min'].append(float(tt[0]) * 2080)
            dict['max'].append(float(tt[1]) * 2080)
    
    elif "a month" in k:
        t = k.replace(" a month", '')
        t = t.replace(",", "")
        tt = np.char.split(t, "–").tolist()
        if len(tt) <2 :
            dict['min'].append(float(tt[0]) * 12)
            dict['max'].append(np.NaN)
        else:
            dict['min'].append(float(tt[0]) * 12)
            dict['max'].append(float(tt[1]) * 12)            

    elif "a year" in k:
        t = k.replace(" a year", '')
        t = t.replace(",", "")
        tt = np.char.split(t, "–").tolist()
        if len(tt) <2 :
            if "K" in tt[0]:
                tt[0] = tt[0].replace("K", '')
                dict['min'].append(float(tt[0]) * 1000)
                dict['max'].append(np.NaN)
            else:
                dict['min'].append(float(tt[0]))
                dict['max'].append(np.NaN)
        else:
            if "K" in tt[0]:
                tt[0] = tt[0].replace("K", '')
                dict['min'].append(float(tt[0]) * 1000)
            if "K" in tt[1]:
                tt[1] = tt[1].replace("K", '')        
                dict['max'].append(float(tt[1]) * 1000)
            else:
                dict['min'].append(float(tt[0]))
                dict['max'].append(float(tt[1]))
```

```python
us['min_salary'] = dict['min']
us['max_salary'] = dict['max']
```
